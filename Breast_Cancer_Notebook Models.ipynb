```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
import sklearn
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('breast-cancer.csv')
```

```python
encoder=preprocessing.LabelEncoder()
df['diagnosis'] = encoder.fit_transform(df['diagnosis'])
df['diagnosis']
```

```python
for i in df.columns:
    print(i," = ",df[i].corr(df['diagnosis']))
```

```python
df.drop(['id','fractal_dimension_mean','texture_se','symmetry_se','perimeter_se','area_se','area_mean','radius_mean'],axis=1,inplace=True)
```

```python
plt.figure(figsize=(25,25))
sb.heatmap(df.corr(), annot=True,cmap='YlGnBu')
```

```python
df.head()
```

```python
for i in df.columns:
    plt.figure(figsize=(5,5))
    sb.boxplot(data=df[i])
```

```python
df.head()
```

```python
X = df.iloc[:,1:]
Y = df.iloc[:,0]
Y
```

```python
col=X.columns
object = StandardScaler()
X = object.fit_transform(X)
```

```python
X=pd.DataFrame(X,columns=col)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42) 
```

```python
model = LogisticRegression(random_state=42)
model = model.fit(x_train, y_train)
model.score(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score( y_test, y_pred))
```

```python
from sklearn.metrics import classification_report
cr=classification_report(y_test, y_pred)
print(cr)
```

```python
from sklearn.metrics import confusion_matrix


cr=confusion_matrix(y_test, y_pred)
cr=pd.DataFrame(cr)
cr.columns = ['M', 'B']
cr.index=['M', 'B']
sb.heatmap(cr, annot=True, annot_kws={"size": 20},cbar=0,xticklabels=True,yticklabels=1, fmt="d",cmap='YlGnBu') 
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=42)
clf = clf.fit(x_train,y_train)
y_pred = clf.predict(x_test)
clf.score(x_train, y_train)
```

```python
print(accuracy_score( y_test, y_pred))
```

```python
from sklearn.metrics import classification_report
cr=classification_report(y_test, y_pred)
print(cr)
```

```python
from sklearn.metrics import confusion_matrix


cr=confusion_matrix(y_test, y_pred)
cr=pd.DataFrame(cr)
cr.columns = ['M', 'B']
cr.index=['M', 'B']
sb.heatmap(cr, annot=True, annot_kws={"size": 20},cbar=0,xticklabels=True,yticklabels=1, fmt="d",cmap='YlGnBu') 
```

```python
from sklearn.ensemble import RandomForestClassifier
Rf = RandomForestClassifier(random_state=42)
Rf = Rf.fit(x_train,y_train)
y_pred =Rf.predict(x_test)
Rf.score(x_train, y_train)
```

```python
print(accuracy_score( y_test, y_pred))
```
